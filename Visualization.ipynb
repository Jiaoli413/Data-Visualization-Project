```python
data.loc[:,'Start_Time'] = pd.to_datetime(data.loc[:,'Start_Time'], format = '%d/%m/%y %H:%M')
```

```python
def DateSplit(df, col):
    year, month, day, time = [], [], [], []
    data = df.loc[:, col].values
    
    for i in range(data.shape[0]):
        year.append(int(str(data[i])[:4]))
        month.append(int(str(data[i])[5:7]))
        day.append(int(str(data[i])[8:10]))
        time.append(str(data[i])[11:16])
    date = pd.DataFrame({'Year': year, 'Month': month, 'Day': day, 'Time': time})
    result = pd.concat([date, df], axis=1)
    return result

data = DateSplit(df=data, col='Start_Time')
```
