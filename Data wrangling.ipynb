```python
import os
import pandas as pd

def file_split(filename, file_num):
    chunksize = 1000000   
    data1 = pd.read_csv(filename, chunksize = chunksize, sep=',', encoding='utf-8') 
    num = 0
    for chunk in data1:
        num += len(chunk)
    chunksize = round(num / file_num + 1)

    head, tail = os.path.splitext(filename)
    data2 = pd.read_csv(filename, chunksize = chunksize, sep=',', encoding='utf-8')
    i = 0 
    for chunk in data2:
        chunk.to_csv('{0}_{1}{2}'.format(head, i, tail), index=False)
        print('The number of {0} data saved...'.format(i))
        i += 1
            

if __name__ == '__main__':
    filename = '../R_working_path/US_accident_dataset/US_Accidents_Dec19.csv'
    file_split(filename, 5)     
```

```python
def DateSplit(df, col):
    year, month, day, time = [], [], [], []
    data = df.loc[:, col].values
    
    for i in range(data.shape[0]):
        year.append(int(str(data[i])[:4]))
        month.append(int(str(data[i])[5:7]))
        day.append(int(str(data[i])[8:10]))
        time.append(str(data[i])[11:16])
    date = pd.DataFrame({'Year': year, 'Month': month, 'Day': day, 'Time': time})
    result = pd.concat([date, df], axis=1)
    return result

data = DateSplit(df=data, col='Start_Time')
```
